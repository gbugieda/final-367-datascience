## **Background**

One of the most devastating and infamous oil spills in recent history was the 2010 Deepwater Horizon oil spill in the Gulf of Mexico.  On April 20, 2010, there was an explosion and gas blowout on the Deepwater Horizon drilling rig.  This explosion resulted in an oil leak that continued polluting the water for 87 days until the well was capped [[2]](https://www.sciencedirect.com/science/article/pii/S0025326X16304313?casa_token=0-93hNxgWtMAAAAA:rek6bK4HdPZaqWW_CJnR8aGsQcr_IJ1jdxjrD55UJ_DkR03-Y8LZYnE5sQjawXEwMBONv_b956k).  All in all, over "3.19 million barrels of oil were released into the ocean during the spill, in addition to several hundred thousand tons of hydrocarbon gases" [[2]](https://www.sciencedirect.com/science/article/pii/S0025326X16304313?casa_token=0-93hNxgWtMAAAAA:rek6bK4HdPZaqWW_CJnR8aGsQcr_IJ1jdxjrD55UJ_DkR03-Y8LZYnE5sQjawXEwMBONv_b956k).  Oil spills and pollution in the ocean can have devastating effects on marine life and coastal ecosystems.  For example, when oil spills into the sea, it "initially spreads over the water surface as a slick a few millimeters thick" [[1]](https://www.sciencedirect.com/science/article/pii/S1353256102000518?casa_token=Obpj-RU2YkYAAAAA:f82TkBy0bmAhOFMvoY8sBQLA-k9yJ9Vl-T12rsh9LjJy25X5SOHyIDjji6mQRwrS5LU-D6Jelwg). In the case of the Deepwater Horizon accident, ocean currents carried oil away from the site of the explosion towards the coasts and beaches and large cleanup operations were implemented to contain and remove the oil as well as to help the marine ecosystems recover.

## **Justification**
* **Question 1:**
   * According to researchers studying bird mortality from the Deepwater Horizon oil spill, "following large oil spills at sea, seabirds are particularly sensitive to both internal and external oil exposure," and their habits "lead to frequent contact with surface oil [[3]](https://www.int-res.com/articles/meps_oa/m513p225.pdf)".  After studying the mortality rates of birds, they found that between "36,000 and 67,000 birds died in the offshore waters of the Gulf of Mexico" due to the Deepwater oil spill [[3]](https://www.int-res.com/articles/meps_oa/m513p225.pdf).  While this is a very large number of seabird deaths, it brings to light the question of how many birds were affected by the oil spill and managed to survive, and what exactly happened to them in terms of oil exposure.  Oil spills can threaten birds in non-lethal ways, with "external contamination of feathers considered the single most harmful effect of oil on birds" because it can lead to loss of water repellency, insulation, and flight functions in feathers [[4]](https://www.sciencedirect.com/science/article/pii/S0147651317303068?casa_token=1afYOOyVXR0AAAAA:iDJs1C45SWuFlv0DGCWr8082J3Te4KgNXAZRV84qQ_mOpID9FgduJ9vdpS-aUWLC4A5SS1dwWDc).  Therefore, it is not only important to study these effects on seabirds after oil spills, but it is also important to look at how far away from the original spill location birds experienced these effects in order to help with recovery efforts.  
* **Question 2:**
   * While most oil spills directly affect marine life on the surface and shore, due to the circumstances of the Deepwater Horizon spill, and the depth at which it occurred, marine life in deeper parts of the ocean were likely impacted as well [[5]](https://cdnsciencepub.com/doi/full/10.1139/f2011-171).  This extensive damage to organisms in all parts of the ocean is especially detrimental to commercial and recreational fishing, as losses would be in the billions [[5]](https://cdnsciencepub.com/doi/full/10.1139/f2011-171). "Impacts on marine ecosystems translate into impacts and the economy and livelihoods, including commercial fisheries, recreations, mariculture, tourism, and energy markets.[[5]](https://cdnsciencepub.com/doi/full/10.1139/f2011-171)"  For example, food safety concerns are often raised regarding fish caught in oil-contaminated areas and neighboring locations, leading to a decrease in the ability to sell fish at previous rates [[5]](https://cdnsciencepub.com/doi/full/10.1139/f2011-171).  "Immediately following the DH spill, the region's key shrimp and oyster fishing areas were officially closed," which was especially significant because "about 70% of the commercially caught shrimp and oysters in the US come from the GOM" (Gulf of Mexico) [[5]](https://cdnsciencepub.com/doi/full/10.1139/f2011-171).  Therefore, oil spills and their lasting effects can harm fishing industries and economies that rely on seafood as their major export.  It is important to look at how these effects can hurt these industries, as well as the resilience of these economies and communities as they attempt to bounce back from near collapse.     
* **Question 3:**
   * There have been several studies on the effects of oil on fish species, and some of the notable effects include "premature hatching, reduced growth rates, genetic abnormalities, increased mortality" as well as increased disease rates [[5]](https://cdnsciencepub.com/doi/full/10.1139/f2011-171).  Oil can also make its way up with food chain, as the oil and hydrocarbons absorbed by plankton, mussels, and other benthic organisms can contaminate larger fish and predatory species that eat these animals for food [[5]](https://cdnsciencepub.com/doi/full/10.1139/f2011-171).  When investigating the impact of oil on fish, evidence from previous oil spills is important to consider.  Studies on salmon population after fish eggs were exposed to the Exxon Oil Spill demonstrated that there was a reduction in the overall production of adult fish, and even low levels of oil poisoning were dangerous because of the delayed reproductive effects [[6]](https://www.tandfonline.com/doi/pdf/10.1080/00908310390195589?needAccess=true).  Therefore, it is important to look at how oil affects fish, specifically in relation to the Deepwater Horizon event to see if any connections can be made between past oil spills or if any new knowledge can be obtained about the resiliency or mortality of fish species affected by oil.

#### Cleaning Data on Capture Sites

```python
siteEffects =  bird_health_df[['Site Oiling', 'Visual Oiling', 'Fluorescence Assessment']]
siteEffects.rename(columns={'Visual Oiling':'visual', 'Fluorescence Assessment': 'fluorescence','Site Oiling':'site'},inplace=True)
# Eliminating any rows where the oil at the site wasn't assessed
siteEffects = siteEffects[siteEffects['site'] != 'No recorded assessment of oiling at site.']

# Shortening words in columns
siteEffects['site'] = siteEffects['site'].apply(lambda x: x.split(' ')[0])
siteEffects['visual'] = siteEffects['visual'].apply(lambda x: x.split(' ')[0])
siteEffects['fluorescence'] = siteEffects['fluorescence'].apply(lambda x: x.split(' ')[0])

siteEffects.head()
```

```python
# Numbering oiling and flurescence
# Calculating how many sites were not assesed for each observation to later subtract
noAssessmentVisual = (siteEffects['visual'] == 'Not').sum()
noAssessmentFluorescent = (siteEffects['fluorescence'] == 'Not').sum()
countVisual = siteEffects['visual'].count()
countFluorescent =siteEffects['fluorescence'].count()

# Dictionaries to change strings to numbers to accurately calculate percents
oilNums = {'No':0.0, 'Not':0.0, 'Trace':1.0, 'Light':1.0, 'Moderate':1.0,'Heavy':1.0}
siteCheck = {'Oiling':'Yes', 'No': 'No'}

# Turning strings into numbers (or for site column, yes or no)
siteEffects.visual = [oilNums[item] for item in siteEffects.visual]
siteEffects.fluorescence = [oilNums[item] for item in siteEffects.fluorescence]
siteEffects.site = [siteCheck[item] for item in siteEffects.site]

siteEffects.head()
```

Since I want to look at the total number of bids observed with oil both visually and with fluorescent lighting, I will need to consolidate the data with sums.

```python
siteEffects = siteEffects.groupby('site').agg({'visual':'sum','fluorescence':'sum'})
print("Total birds surveyed at sites visually: " + str(countVisual - noAssessmentVisual))
print("Total birds surveyed at sites fluorescentally: " + str(countFluorescent - noAssessmentFluorescent))
print("Chart of number of birds with oil at each site below:")
siteEffects.head()
```

Now, I'll plot the data to create a visualization.

```python
siteEffects.plot.bar(rot=0,color={'visual':'lightskyblue','fluorescence':'mediumslateblue'},figsize=(10,7))
plt.suptitle("Number of Birds That Were Covered In Oil at Various Locations")
plt.title("Observations Made Visually and With UV Light")
plt.xlabel("Site Contained Oil")
plt.ylabel("Number of Oil-Covered Birds")

plt.show()
```

From the barchart above, the data shows that even when birds may not visually look like they have oil on them, further examination using fluorescent lights can reveal an impact of oil on bird feathers.  Therefore, the effects of oily residue can't always be seen easily.  Even when the sites where the birds were captured didn't contain oil, oil was still found on the animals, alluding to the fact that oil from an oil spill can find its way to wildlife indirectly.  Birds travel and migrate, and they can pick up oil that stays with them even as they move farther away from the impact site.

### **Effects of Oil Spills on Local Fishing and Coastal Economies**

#### Investigating how much oil remains in the sand after an oil spill:

One major question about the lasting impacts of oil spills is, where does all the oil go?  One of the places oil ends up is in sand, especially in the coastal beaches around the oil spill origination.  This is important for coastal beaches to consider, especially because their tourism and industries could be affected by the remnants of oil in the sand. I want to see if the oil stays in the sand over time, or if the situation clears up after a few years.

To start, I'll read in a dataset that contains information about the oil content in sand patties.  This data contains oil and moisture sand content information from Alabama, Florida, Mississippi, and Louisiana.

```python
oil_sand_df = pd.read_excel('data/oil-sand.xlsx',sheet_name="data")
## need to drop last 4 rows due to non-data information attached by source
oil_sand_df.drop(oil_sand_df.tail(4).index,inplace = True) ## dropping last 4 rows
oil_sand_df.tail() 
```

```python
#change location to just state name and let's plot the oil content over time 
oil_areas_df = oil_sand_df
oil_areas_df['Sampling Location'] = oil_sand_df['Sampling Location'].apply(lambda x: x.split(',')[1].strip())
oil_areas_df.head(2)
```

Now, I'll plot the data over time so we can see the oil content over time between the different locations. In terms of the data, I'll be changing fractional oil content to the percentage of oil in the sand for easier analysis interpretation.

```python
plt.subplots(figsize=(8,5))
oil_areas_df['Fractional Oil Content'] = oil_sand_df['Fractional Oil Content'].apply(lambda x: x * 100)
ax = sns.lineplot(data=oil_areas_df, x='Collection Date', y='Fractional Oil Content', hue='Sampling Location',estimator=np.median,ci=None,palette="hls") 
plt.title('Percentage of Oil Content in Sand Patties Over Time',fontsize=15)
plt.ylabel("Oil Content (%)",fontsize=12)
plt.xlabel("Collection Date",fontsize=12)
xticks = ax.xaxis.get_major_ticks()
for i in range(len(xticks)):
    if i % 8 == 1:
        xticks[i].set_visible(True)
    else:
        xticks[i].set_visible(False)
plt.show()
```

```python
## Measuring the Kendall Tau correlation coefficient between time and oil content
## This will determine if there is a monotnous relationship between the two variables

tau, p_value = stats.kendalltau(oil_areas_df['Collection Date'], oil_areas_df['Fractional Oil Content'])
print("Kendall Tau Correlation Coefficient: " + str(tau))
if p_value < 0.05:
    print("Since the pvalue is less than 0.05, we reject the null hypothesis that there is no assocation.\n")
else:
    print("Since the pvalue is greater than 0.05, we accept the null hypothesis that there is no association.\n")
```

As seen in the lineplot above, despite the fact that the Deepwater Horizon incident took place in 2010, even years later, in 2014, the sand still had an oil content of around 10-20% in all locations.  Although there were some increases and decreases in the amount of oil in the sand over time, it never completely disappeared.  Even as the sand endured weathering, portions of oil still remained.  Even more concerning is the fact that the Kendall Tau statistical test showed no association between time and oil content, meaning that after an oil spill, the amount of oil in the sand doesn't produce a monotonic relationship.  While it is good that there was no positive correlation, the downside is that there is no negative correlation either; this means the evidence does not clearly show a decrease in oil content over time.  Knowing that oil residue is pervasive and persistent, it brings up concerns about wildlife and even human activity on the beaches affected by oil spills.

### **Effects of Oil Spills on Fish**

One of the concerns after oil spills is if the excess production of methylmercury in the ocean can be picked up by fish, leading to higher MeHg levels.  Not only could high mercury levels in fish harm the fish themselves, but it could also impact whether or not the fish is deemed viable for consumption or seafood sale.  To investigate this topic, I want to see if there is a difference between the levels of MeHg in different species of fish after and oil spill and whether or not the fish contained unsafe levels of MeHg in their bodies.

```python
fish_df = pd.read_excel('data/fish.xlsx',sheet_name="Sheet1", header=2) #need to get in headers for columns when reading in data
fish_df.drop(fish_df.head(1).index,inplace = True) #removing column description row
fish_df.drop(list(fish_df.filter(regex = 'Unnamed')), axis = 1, inplace = True)
fish_df.head()
```

There were some issues with how the fish names were inputted into this xlsx file, so I did some renaming and dropping to match the species metadata associated with this dataset:

```python
fish_df['Species'] = fish_df['Species'].str.lower()    
updateFish = {'rock seabass':'rock sea bass','bank seabass':'bank sea bass','albaco jack':'almaco jack','amberjack liver':'amberjack','pigfish':'pinfish'}

fish_df.update(fish_df['Species'].replace(updateFish))
fish_df = fish_df[fish_df.Species != 'grouper']
fish_df = fish_df[fish_df.Species != 'eel liver']
fish_df = fish_df[fish_df.Species != 'red grouper']

mercury = fish_df[['Species','MeHg']]
mercury = mercury.dropna()
mercury.head()
```

In my analysis, I'll be looking at the MeHg, or methylmercury levels, in the fish.  I will later be comparing MeHg levels to what is declared safe for consumption by the FDA.  The FDA recommendation is in ppm (parts per million) instead of the ppb (parts per billion) listing in this dataset.  Therefore, I'll make the MeHg column into a measure of ppm:

```python
mercury['MeHg'] = mercury['MeHg'].apply(lambda x: round(x/1000,2))
mercury.head(2)
```

Since I want to run some statistical analysis on the data, I need to figure out if the population is normal.  To do that, I'll run a shapiro test and plot a histogram of the data.

```python
## CHECKING NORMALITY WITH HISTOGRAM
mercury['MeHg'].hist(color = "skyblue")
plt.title("Distribution of MeHg Levels")
plt.xlabel("MeHg Levels (ppm)")
plt.ylabel("Frequency")
plt.show()

## CHECKING NORMALITY WITH SHAPIRO TEST
test_statistic, p_value = stats.shapiro(mercury['MeHg'])
if p_value < 0.05:
    print("Since the pvalue is less than 0.05, we reject the null hypothesis that the MeHg levels are normally distributed.\n")
else:
    print("Since the pvalue is greater than 0.05, we accept the null hypothesis that the MeHg levels are normally distributed.\n")
```

Based on this test and the histogram, we know that the distribution is NOT normal.  Since the distribution is skewed right, we can't use any statistics tests assuming normality.  Therefore, let's use kruskal wallis test to see what we can find:

```python
statistic, p_value = stats.kruskal(*[group['MeHg'].values for name,group in mercury.groupby('Species')])
            
print("Kruskal-Wallis H-Test to see if population median of different fish species MeHg levels are equal:\n")
print("Test Statistic: " + str(statistic) + ", P-Value: " + str(p_value) + ", P-Value < 0.05?: " + str(p_value < .05))
if p_value < 0.05:
    print("Since the pvalue is less than 0.05, we reject the null hypothesis. \n")
else:
    print("Since the pvalue is greater than 0.05, we accept the null hypothesis.\n")
```

Based on the results of the Kruskal test, since the p-value < alpha of 0.05, we reject the null hypothesis that the population medians of all of the fish species are the same.  There is sufficient evidence to suggest that the median MeHg levels are different.  This means that as a result of increased methlymercury in the ocean, some fish species were more susceptible to picking it up than others.  Let's see which species picked up the most, and which species are over the mercury level standard given by the FDA, which is 1ppm:

```python
medianMercury = mercury.groupby('Species').agg({'MeHg':'median'}).reset_index().sort_values('MeHg',ascending = False)

plt.subplots(figsize=(12,8))
sns.set_style("whitegrid")
sns.boxplot(data=mercury,x="MeHg",y='Species',orient='h',order=medianMercury['Species'])

plt.title("Distribution of Methylmercury in Different Fish Species",fontsize=16)
plt.xlabel("Level of MeHg (ppm)",fontsize=14)
plt.ylabel("Marine Species",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

plt.subplots(figsize=(12,8))
sns.set_style("whitegrid")
sns.barplot(data=medianMercury,x="MeHg",estimator=np.median,y='Species',orient='h',ci=None)

plt.title("Median Levels of Methylmercury in Different Fish Species",fontsize=16)
plt.xlabel("Level of MeHg (ppm)",fontsize=14)
plt.ylabel("Marine Species",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

From the boxplots, we can see that cutthroat eels and king snake eels not only have the largest range of MeHg values, they also have the most significant outliers as well, with some containing over 12.5 ppm of MeHg. We can also see that shrimp, hermit crabs, tomtates, and grey triggerfish have some of the smallest variations of MeHg levels, and all values observed during the study seem to be within the range of FDA safety guidelines for ppm levels. 

From the bar chart, we can see that 15 out of 26 marine species had median levels of methylmercury that would be in the FDA's "deemed safe" range for consumption.  That means that 11 species had levels higher than that range, with the cutthroat eel having the highest median mercury levels out of any other species. High levels of MeHg ppm are not good for humans, and they are also not good for fish, highlighting this unfortunate effect of oil spills: the increased methylmercury in the ocean does get absorbed by fish species (some at higher rates than others) which is a serious health concern.

I found that there were birds affected by oil at locations hundreds of miles away from the original oil spill.  In addition, I was able to show that even when birds don't look like they have oil on them, the numbers of birds affected might be much higher because fluorescent light shows residue that can't otherwise be visually ascertained.  
* **How do oil spills affect local fishing and coastal economies?**
    * I looked at the oil content in sand along the coasts after the Deepwater Horizon incident, and discovered that the oil doesn't just go away over time.  Even after a few years post-spill, 10-20% oil content remained in the sand.  Areas that once relied on their picturesque and clean beaches for tourism and recreation activities might lose out to other areas that weren't affected by the oil.  Another industry that is affected is the fishing industry, where the results of data I analyzed showed a drop in the value and landings of four types of seafood all during 2010, the year of the spill. 
* **What is the effect of oil on fish species?**
    * Looking closer at the fish and marine life affected by oil spills, especially those that are fished and commercially distributed, I found that some species absorbed more MeHg than others.  High levels of MeHg in fish species is dangerous not only to the fish, but to humans who might later consume the fish after commercial distribution.

All in all, my analysis showed the harmful effects of Deepwater Horizon oil spill on the Gulf of Mexico ecosystems, as well as general detriments to marine life and human activities that could occur in future ocean oil disasters.

## **Ethical Concerns:**

* Stakeholder: Offshore drilling companies
     * Data released about the effects of oil spills on marine life and ecosystems could lead to company profit loss if offshore drilling companies are no longer utilized because of fears over more oil spills.  There could also be increased negative public opinion of these companies with the publicity of oil-related animal deaths.
* Stakeholder: Tourism Businesses
     * Data released about places still recovering from oil spills could deter tourists from using beaches or swimming on shorelines that are still affected by residual oil or oil damages.  This could lead to a decrease in profits for coastal communities that rely on tourism to survive.
* Stakeholder: Fishing Industry
     * Fishing industries along coastal communities affected by oil spills could be hurt by inability to catch/sell fish (because of increase in dead fish due to the oil) as well as public distrust of buying seafood from area inundated with crude oil. Also, due to the threat of MeHg levels in fish after an oil spill, food safety regulations could get harsher and commercial distributors might not be able to profit off of all of the fish they caught.
